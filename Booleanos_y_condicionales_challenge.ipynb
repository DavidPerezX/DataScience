1.-Escriba un programa que pida dos números enteros y que calcule su división, escribiendo si la división es exacta o no.

```python
terminado = False
while terminado == False:
  print("---Calculadora que divide dos números---")
  primerN=int(input("Introduce el primer número: "))
  segundoN=int(input("Introduce el segundo número: "))
  calculo=primerN/segundoN
  if (primerN%segundoN==0):
    print(f"El resultado es: {calculo} y la división es exacta")
  else:
    print(f"El resultado es: {calculo} y la división es inexacta")
  opcion =input('Para terminar ingresa 0 y para continuar cualquier tecla: ')
  if opcion == "0":
    terminado = True
    print("Bye!")

```

2.-Escriba un programa que pida dos números y que conteste cuál es el menor y cuál el mayor o que escriba que son iguales.

```python
terminado = False
while terminado == False:
  print("---Comparador de números---")
  primerN=int(input("Introduce el primer número: "))
  segundoN=int(input("Introduce el segundo número: "))
  if (primerN>segundoN):
    print(f"{primerN} es mayor que {segundoN}")
  elif(segundoN>primerN):
    print(f"{segundoN} es mayor que {primerN}")
  else:
    print(f"Los números son iguales ")
  opcion =input('Para terminar ingresa 0 y para continuar cualquier tecla: ')
  if opcion == "0":
    terminado = True
    print("Bye!")
```

3.-Escriba un programa que pida dos números enteros y que escriba si el mayor es múltiplo del menor.

```python
terminado = False
while terminado == False:
  print("---Revisa mayor y múltiplo---")
  primerN=int(input("Introduce el primer número: "))
  segundoN=int(input("Introduce el segundo número: "))
  if (primerN>segundoN):
    if(primerN%segundoN==0):
      print(f"{primerN} es mayor que y es multiplo de {segundoN}")
    else:
      print(f"{primerN} es mayor que {segundoN} y NO es multiplo de {segundoN}")
  elif (segundoN>primerN):
    if(segundoN%primerN==0):
      print(f"{segundoN} es mayor que y es multiplo de {primerN}")
    else:
      print(f"{segundoN} es mayor que {primerN} y NO es multiplo de {primerN}")
  else:
    print(f"Los números son iguales ")
  opcion =input('Para terminar ingresa 0 y para continuar cualquier tecla: ')
  if opcion == "0":
    terminado = True
    print("Bye!")
```

4.-Escriba un programa que pregunte primero si se quiere calcular el área de un triángulo o la de un círculo. Si se contesta que se quiere calcular el área de un triángulo (escribiendo T o t), el programa tiene que pedir entonces la base y la altura y escribir el área. Si se contesta que se quiere calcular el área de un círculo (escribiendo C o c), el programa tiene que pedir entonces el radio y escribir el área.

Se recuerda que el área de un triángulo es base por altura dividido por 2 y que el área de un círculo es Pi (aproximadamente 3,141592) por el radio al cuadrado.

```python
terminado = False

while terminado == False:
    eleccion = input('¿A qué figura deseas calcularle el area? T o C:')
    if eleccion =='T':
        base=float(input('Digita la base del triangulo:'))
        print ('Digita la altura del triangulo: ')
        altura=float(input())
        at=(base*altura)/2
        print ('El area del triangulo es: %s'% str(at))
    elif eleccion=='C':
        radio=float(input('Digita el radio del circulo: '))
        ac=3.14*(radio**2)
        print ('El area del circulo es: %s' % str(ac))
    opcion =input('Para terminar ingresa 0 y para continuar cualquier tecla: ')
    if opcion == "0":
      terminado = True
      print("Bye!")
```

Calcular el Índice de masa corporal

Con base en el resultado del imc, a través de un condicional, mostrar la categoría a la que pertenece

```python
terminado = False
while terminado == False:
  print("---Calculadora IMC---")
  alturaf=float(input("Introduce tu altura en centimetros: "))
  altura=alturaf*0.01
  peso=float(input("Introduce tu peso: "))
  imc=(peso/(altura*altura))
  print(f"Tu IMC es: {imc} \nTu estado es: ")
  if (imc<=18.49):
    print("Insuficiencia ponderal/Bajo peso")
  elif (imc>=18.5 and imc<24.9):
    print(f"Normopeso")
  elif (imc>=25 and imc<29.9):
    print(f"Sobrepeso")
  elif (imc>=30 and imc<34.9):
    print(f"Obesidad grado I")
  elif (imc>=35 and imc<39.9):
    print(f"Obesidad grado II")
  elif (imc>=40):
    print(f"Obesidad grado III morbida")
  else:
    print(f"Números inválidos ")
  opcion =input('Para terminar ingresa 0 y para continuar cualquier tecla: ')
  if opcion == "0":
    terminado = True
    print("Bye!")
```
